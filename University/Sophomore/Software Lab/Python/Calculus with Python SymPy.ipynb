1. For any principal amount $p$, any rate $r$, and any number of years $t$, the 
compound interest is calculated using the formula:
$$ A = P\left(1+\frac{r}{n}\right)^{nt}$$ 

Assume you are 20 year old now, you deposited 1 Million bath to bank this year. If you wait until retirement, interest paid __Semiannually__ for 8%.

How much AMOUNT in million bath you could widthdraw at after 60 year old?

```python
from sympy import symbols, evalf, plot
P, n, r, t,x= symbols('P,n,r,t,x')
A = P*(1+(r/n))**(n*t)
print(A.subs({P:1e6, n:2 , r:0.08 , t:40}))
A = A.subs({P:1e6, n:2 , r:0.08 })
plot(A,(t,1, 40))
```

2.  Given $ y= -x^2 + 2x +15 $ plot graph of y(x), y'(x) and solve for maximum value of y(x). 

```python
!pip install matplotlib
```

```python
from matplotlib import pyplot as plt
from sympy import symbols,plot, diff, solve,nsimplify,pprint

x = symbols('x')
y = -x**2+2*x+15
dydx = diff(y)

critical_point = dydx.factor()
critical_point = solve(dydx)

cp = [critical_point[i].evalf(2) for i in range(len(critical_point))]
dydx2 = diff(y,x,2)
extrema = [dydx2.subs(x , cp[i]) for i in range(len(cp))]

plot (y)
plot (dydx)
print(extrema)
print (cp)
```
